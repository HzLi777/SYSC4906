```python
from datasets import load_dataset

train_dataset = load_dataset("cifar100", split="train")
test_dataset = load_dataset("cifar100", split="test")
```

```python
from torchvision import transforms
import torch
from einops import rearrange
import random
from torch.utils.data import TensorDataset, DataLoader
import timm
import math
import time
```

```python
# Prepare data to put in data loaders (this can be done in many different ways, here is one way)

PIL_to_tensor = transforms.ToTensor()
data = []
for row in train_dataset:
    img = PIL_to_tensor(row['img'])  # (3, 32, 32)
    label = torch.LongTensor([row['fine_label']])  # (1)

    # add a leading dimension so we can stack images along it
    img = rearrange(img, 'c h w -> 1 c h w')  # (1, 3, 32, 32)

    # append to list
    data.append([img, label])

random.shuffle(data)  # shuffle data before split
train_data = data[:40_000]
val_data = data[40_000:]

train_images = torch.cat([x[0] for x in train_data], dim=0)  # (40_000, 3, 32, 32)
train_labels = torch.cat([x[1] for x in train_data], dim=0)  # (40_000)

val_images = torch.cat([x[0] for x in val_data], dim=0)  # (10_000, 3, 32, 32)
val_labels = torch.cat([x[1] for x in val_data], dim=0)  # (10_000)
```

```python
# Double check values are reasonable and datatypes are what we expect. In this case we expect normalized pixel values that are floats. Images are often stored as 8-bit integers, so this is an important step

print(train_images[0, :, :5, :5])
```

```python
# We should choose the biggest batch size that will fit on our GPU
# You will get an out-of-memory error if your batch size is too big, but you don't need to search for the biggest possible batch size—this would take a while. As long as your GPU memory is almost full, you're good.

batch_size = 256  # arbitrarily chosen for now
train_loader = TensorDataset(train_images, train_labels)
train_loader = DataLoader(train_loader, batch_size=batch_size, shuffle=True)

val_loader = TensorDataset(val_images, val_labels)
val_loader = DataLoader(val_loader, batch_size=batch_size, shuffle=False)
```

```python
train_labels.shape
```

```python
base_lr = 1e-4
learning_rate = base_lr * batch_size / 256  # nice rule-of-thumb to use

model = timm.create_model('resnet50d', num_classes=100, pretrained=True)
model.cuda()  # put model on GPU
opt = torch.optim.AdamW(model.parameters(), lr=learning_rate)  # AdamW is a often used
loss_function = torch.nn.CrossEntropyLoss()  # CE Loss because we have 100 classes (were only 1 is correct)
epochs = 10  # arbitrarily chosen for now

for epoch in range(epochs):
    start_time = time.time()
    train_losses = []
    model.train()  # put model in "train" mode
    for batch in train_loader:
        batch_imgs, batch_labels = batch
        batch_imgs = batch_imgs.cuda()  # (bsz, 3, 32, 32), this puts our images on the GPU
        batch_labels = batch_labels.cuda()  # (bsz), this puts our labels on the GPU

        logits = model(batch_imgs)  # (bsz, 100), this sends our data through out model, returning the class logits (this is a "forward pass")
        loss = loss_function(logits, batch_labels)  # compute the CE loss between logits and labels (internally does a softmax)
        loss.backward() # computes the gradients of every parameter (this is a "backward pass")

        opt.step()  # updates parameters based on the gradients (and their momentums in this case) and learning rate
        opt.zero_grad()  # resets gradients
        train_losses.append(loss.item())

    val_losses = []
    model.eval()  # put model in "eval" mode
    with torch.no_grad():  # when evaluating, we don't need gradients
        for batch in val_loader:
            batch_imgs, batch_labels = batch
            batch_imgs = batch_imgs.cuda()  # (bsz, 3, 32, 32)
            batch_labels = batch_labels.cuda()  # (bsz)

            logits = model(batch_imgs)  # (bsz, 100)
            loss = loss_function(logits, batch_labels)
            val_losses.append(loss.item())

    train_time = time.time() - start_time
    epoch_train_loss = torch.Tensor(train_losses).mean().item()
    epoch_val_loss = torch.Tensor(val_losses).mean().item()
    print(f'Epoch: {epoch}  Train Loss: {epoch_train_loss:8.6f}   Val Loss: {epoch_val_loss:8.6f}  Time: {train_time:6.4f}')

```

```python
# This is a common learning rate schedule, we haven't covered it yet

def adjust_learning_rate(epoch, sched_config):
    """Decay the learning rate with half-cycle cosine after warmup"""
    if epoch < sched_config['warmup_epochs']:
        lr = sched_config['lr'] * epoch / sched_config['warmup_epochs']
    else:
        lr = sched_config['min_lr'] + (sched_config['lr'] - sched_config['min_lr']) * 0.5 * \
            (1. + math.cos(math.pi * (epoch - sched_config['warmup_epochs']) / (sched_config['epochs'] - sched_config['warmup_epochs'])))
    return lr

def get_lr(optimizer):
        for param_group in optimizer.param_groups:
            return param_group['lr']

sched_config = {'lr': learning_rate,
                'warmup_epochs': 5,
                'min_lr': 0,
                'epochs': epochs}

set_lr = adjust_learning_rate(epoch, sched_config)  # get LR for this epoch

for g in opt.param_groups:
    g['lr'] = set_lr  # update
current_lr = get_lr(opt)  # confirm
```
